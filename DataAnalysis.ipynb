```python
# Count job postings by State.
location_counts = df['State'].value_counts()

# Display the distribution of job postings by location
print("Job Postings Distribution by State:")
print(location_counts)
```

```python
# Analyze job distribution across different State using bar chart.

plt.figure(figsize=(12, 6))
sns.countplot(y='State', data=df, order=df['State'].value_counts().index)
plt.title('Distribution of Locations')
plt.show()
```

```python
# Count job postings by Industry.
Industry_counts = df['Industry'].value_counts()

# Display the distribution of job postings by location
print("Job Postings Distribution by Industry:")
print(location_counts)
```

```python
# Analyze trends in industries with the highest number of job postings using bar chart.

plt.figure(figsize=(12, 6))
sns.countplot(y='Industry', data=df, order=df['Industry'].value_counts().index)
plt.title('Distribution of Industries')
plt.show()
```

```python
# Analyze the frequency count of industries using pie chart
industry_counts = df['Industry'].value_counts()[:20]

# Percentage analysis
industry_percentages = (industry_counts / len(df)) * 100

# Combine counts and percentages into a DataFrame
industry_distribution = pd.DataFrame({
    'Industry': industry_counts.index,
    'Count': industry_counts.values,
    'Percentage': industry_percentages.values
})

# Display the frequency count and percentage analysis
print(industry_distribution)

# Plot a pie chart
plt.figure(figsize=(10, 8))
plt.pie(industry_distribution['Count'], labels=industry_distribution['Industry'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Postings by Industry')
plt.show()
```

```python
# Analyze trends in industries with the highest number of job postings using bar chart.

plt.figure(figsize=(12, 6))
sns.countplot(y='Industry', data=df, order=df['Industry'].value_counts().index)
plt.title('Distribution of Industries')
plt.tight_layout()
plt.savefig('C:/Users/syaki/Documents/~COURSE WORK/Distribution of Job Posting by Industries.png')
plt.show()
```
