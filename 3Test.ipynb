```python
# Count job postings by State
location_counts = df['State'].value_counts()

# Display the distribution of job postings by location
print("Job Postings Distribution by State:")
print(location_counts)
```

```python
# Analyze trends in industries with the highest number of job postings using bar chart

plt.figure(figsize=(12, 6))
sns.countplot(y='Industry', data=df, order=df['Industry'].value_counts().index)
plt.title('Distribution of Industries')
plt.show()
```

```python
# Explore job distribution across different State using bar chart

plt.figure(figsize=(12, 6))
sns.countplot(y='State', data=df, order=df['State'].value_counts().index)
plt.title('Distribution of Locations')
plt.show()
```

```python
# Analyze No of Job Posting from different Industry
# Using Bar chart showing the distribution of industries.

# Extract the 'Industry' column
industry_data = df['Industry']

# Create a frequency distribution
industry_freq = industry_data.value_counts()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=industry_freq.index, y=industry_freq.values, palette='viridis')
plt.title('Frequency Distribution of Job Postings by Industry')
plt.xlabel('Industry')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
## Analyze the frequency distribution of categorical variables btw each industry
# Bar chart showing the distribution of industries.

# Extract the 'INDUSTRY' column
industry_data = df['Industry']

# Create a frequency distribution
industry_freq = industry_data.value_counts()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=industry_freq.index, y=industry_freq.values, palette='YlGn')
plt.title('Frequency Distribution of Job Postings by Industry')
plt.xlabel('Industry')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Analyze the frequency count of industries using pie chart
industry_counts = df['Industry'].value_counts()[:20]

# Percentage analysis
industry_percentages = (industry_counts / len(df)) * 100

# Combine counts and percentages into a DataFrame
industry_distribution = pd.DataFrame({
    'Industry': industry_counts.index,
    'Count': industry_counts.values,
    'Percentage': industry_percentages.values
})

# Display the frequency count and percentage analysis
print(industry_distribution)

# Plot a pie chart
plt.figure(figsize=(10, 8))
plt.pie(industry_distribution['Count'], labels=industry_distribution['Industry'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Postings by Industry')
plt.show()
```

```python
# Bar chart showing the distribution of industries by percentages

plt.figure(figsize=(12, 6))
sns.barplot(x='Percentage', y='Industry', data=industry_distribution, palette='viridis')
plt.title('Distribution of Industries')
plt.xlabel('Percentage')
plt.ylabel('Industry')
plt.show()
```

```python
# Analyze the frequency count of State using pie chart
state_counts = df['State'].value_counts()

# Percentage analysis
state_percentages = (state_counts / len(df)) * 100

# Combine counts and percentages into a DataFrame
state_distribution = pd.DataFrame({
    'State': state_counts.index,
    'Count': state_counts.values,
    'Percentage': state_percentages.values
})

# Display the frequency count and percentage analysis
print(state_distribution)

# Plot a pie chart
plt.figure(figsize=(10, 8))
plt.pie(state_distribution['Count'], labels=state_distribution['State'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Postings by State')
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
# Concatenate all job skills into a single string
text = ' '.join(df['Job Skills'].dropna())

# Define a custom Word Cloud with unique parameters
wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=200,
                      colormap='viridis', contour_width=1, contour_color='black').generate(text)

# Plot the Word Cloud
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Unique Word Cloud of Job Skills', fontsize=16)
plt.show()
```

```python
wordcloud.to_file('wordcloud_job1_skills.png')
```

```python
# Analyze the employee distribution using bar chart

plt.figure(figsize=(12, 6))
sns.histplot(df['Employees'], bins=20, kde=True, color='orange')
plt.title('Distribution of Employee Size in Companies')
plt.xlabel('Number of Employees')
plt.show()
```

```python
# Analyze the employee distribution using bar chart

plt.figure(figsize=(12, 6))
sns.countplot(y='Employees', data=df, order=df['Employees'].value_counts().index)
plt.title('Numbers of Employees')
plt.show()
```

```python
# Analyze the datewise distribution of Job Postings

plt.figure(figsize=(12, 6))
datewise_distribution = df['Date Post'].value_counts().sort_index()
datewise_distribution.plot(marker='o')
plt.title('Datewise Distribution of Job Postings')
plt.xlabel('Date Post')
plt.ylabel('Number of Job Postings')
plt.gca().invert_xaxis()  # Mirror the x-axis
plt.grid(True)
plt.show()
```
